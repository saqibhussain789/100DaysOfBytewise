```python
import matplotlib.pyplot as plt

# List of values over time
values = [1, 3, 2, 5, 7, 8, 6, 9, 11, 10]

# Create a line plot
plt.plot(values, marker='o')
plt.title('Trend of Values Over Time')
plt.xlabel('Time')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
# Categories and their frequencies
categories = ['A', 'B', 'C', 'D', 'E']
frequencies = [5, 7, 3, 8, 4]

# Create a bar chart
plt.bar(categories, frequencies, color='skyblue')
plt.title('Frequency of Different Categories')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np

# Generate random data
x = np.random.rand(50)
y = np.random.rand(50)

# Create a scatter plot
plt.scatter(x, y, c='red', alpha=0.5)
plt.title('Scatter Plot of Two Variables')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Drop non-numeric columns
numeric_iris = iris.drop(columns=['species'])

# Compute the correlation matrix
corr = numeric_iris.corr()

# Create a heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# First subplot
axs[0, 0].plot(values, marker='o')
axs[0, 0].set_title('Line Plot')

# Second subplot
axs[0, 1].bar(categories, frequencies, color='skyblue')
axs[0, 1].set_title('Bar Chart')

# Third subplot
axs[1, 0].scatter(x, y, c='red', alpha=0.5)
axs[1, 0].set_title('Scatter Plot')

# Fourth subplot
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5, ax=axs[1, 1])
axs[1, 1].set_title('Heatmap')

plt.tight_layout()
plt.show()
```

```python
# Set color palette
sns.set_palette("pastel")

# Create a customized box plot
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.title('Box Plot of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

```python
# Create a violin plot
sns.violinplot(x='species', y='sepal_length', data=iris)
plt.title('Violin Plot of Sepal Length by Species')
plt.show()
```

```python
# Create a histogram with a KDE plot overlay
sns.histplot(iris['sepal_length'], kde=True, color='blue')
plt.title('Histogram and KDE of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns
import numpy as np
import pandas as pd
from scipy import stats

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Calculate mean, median, and mode of sepal lengths
mean_sepal_length = np.mean(iris['sepal_length'])
median_sepal_length = np.median(iris['sepal_length'])
mode_sepal_length = iris['sepal_length'].mode()[0]

print(f"Mean of sepal lengths: {mean_sepal_length}")
print(f"Median of sepal lengths: {median_sepal_length}")
print(f"Mode of sepal lengths: {mode_sepal_length}")
```

```python
# Calculate variance and standard deviation of petal widths
variance_petal_width = np.var(iris['petal_width'])
std_dev_petal_width = np.std(iris['petal_width'])

print(f"Variance of petal widths: {variance_petal_width}")
print(f"Standard deviation of petal widths: {std_dev_petal_width}")
```

```python
import seaborn as sns
import numpy as np
import pandas as pd

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Exclude non-numeric columns
numeric_iris = iris.select_dtypes(include=[np.number])

# Summary table for all numerical features
summary_table = numeric_iris.describe().T
summary_table['median'] = numeric_iris.median()
summary_table['variance'] = numeric_iris.var()
summary_table['std_dev'] = numeric_iris.std()
summary_table = summary_table[['mean', 'median', 'variance', 'std_dev']]

print(summary_table)
```

```python
# Define a random variable for sepal length and calculate probability distribution
sepal_length_counts = iris['sepal_length'].value_counts(normalize=True)
print(sepal_length_counts)
```

```python
import matplotlib.pyplot as plt

# Plot the probability distribution of sepal lengths
plt.hist(iris['sepal_length'], bins=20, density=True, edgecolor='k')
plt.title('Probability Distribution of Sepal Lengths')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Probability Density')
plt.show()
```

```python
# Calculate and plot CDF for petal lengths
sorted_petal_lengths = np.sort(iris['petal_length'])
cdf = np.arange(len(sorted_petal_lengths)) / float(len(sorted_petal_lengths))

plt.plot(sorted_petal_lengths, cdf)
plt.title('Cumulative Distribution Function (CDF) of Petal Lengths')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Cumulative Probability')
plt.show()
```

```python
# Calculate and plot PDF for sepal width
sepal_width_density = stats.gaussian_kde(iris['sepal_width'])

x = np.linspace(min(iris['sepal_width']), max(iris['sepal_width']), 1000)
plt.plot(x, sepal_width_density(x))
plt.title('Probability Density Function (PDF) of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Density')
plt.show()
```

```python
# Given value
value = 5.0

# Calculate probability
probability = np.mean(iris['petal_length'] > value)
print(f"Probability of petal length greater than {value} cm: {probability}")
```

```python
# Perform a t-test to compare means of petal lengths between two species
species1 = iris[iris['species'] == 'setosa']['petal_length']
species2 = iris[iris['species'] == 'versicolor']['petal_length']

t_stat, p_value = stats.ttest_ind(species1, species2)
print(f"T-statistic: {t_stat}, P-value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("There is a significant difference in mean petal length between the two species.")
else:
    print("There is no significant difference in mean petal length between the two species.")
```

```python
# Calculate covariance and correlation
covariance = np.cov(iris['sepal_length'], iris['sepal_width'])[0, 1]
correlation = np.corrcoef(iris['sepal_length'], iris['sepal_width'])[0, 1]

print(f"Covariance between sepal length and sepal width: {covariance}")
print(f"Correlation between sepal length and sepal width: {correlation}")

# Interpretation
if correlation > 0:
    print("There is a positive correlation between sepal length and sepal width.")
elif correlation < 0:
    print("There is a negative correlation between sepal length and sepal width.")
else:
    print("There is no correlation between sepal length and sepal width.")
```
